```python
def create_download_link_csv(df, title = "Download CSV file", filename = "data.csv"):  
    # generate in-memory CSV, then base64-encode it
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

## Création des differents groupes

```python
for i in range(0,7):
    nombreMois1 = "Nombre_de_mois_"+str(groupes[i])
    nombreMois2 = "Nombre_de_mois_"+str(groupes[i+1])
    nombreMois3 = "Nombre_de_mois_"+str(groupes[i+2])
    valeurAjout1 = "valeur_ajout_"+str(groupes[i])
    valeurAjout2 = "valeur_ajout_"+str(groupes[i+1])
    valeurAjout3 = "valeur_ajout_"+str(groupes[i+2])
    capitauxPropre1 = "capitaux_propr_"+str(groupes[i])
    capitauxPropre2 = "capitaux_propr_"+str(groupes[i+1])
    capitauxPropre3 = "capitaux_propr_"+str(groupes[i+2]) 
    
    #print(nombreMois1,nombreMois2,nombreMois3,valeurAjout1,valeurAjout2,valeurAjout3,capitauxPropre1,capitauxPropre2,capitauxPropre3)
    df_groupe = df_data_1.loc[ (df_data_1[nombreMois1] == 12)& \
                             (df_data_1[nombreMois2] == 12)&(df_data_1[nombreMois3] == 12)&\
                             (df_data_1[valeurAjout1]> 0 )&(df_data_1[valeurAjout2]> 0 )&\
                              (df_data_1[valeurAjout3]> 0 ) & (df_data_1[capitauxPropre1]> 0 )&\
                             (df_data_1[capitauxPropre2]> 0 )&(df_data_1[capitauxPropre3]> 0 )&
                             ((df_data_1["Situation_juridique"] == "Situation normale") | (df_data_1["Situation_juridique"] == "Non précisé")) ]
    df.append(df_groupe)
```

```python
for i in range(0,7):
    print("le groupe ", i+1, "contient les années",groupes[i]," à ",groupes[i+2],"et comporte",df[i].shape,"observations")
```

## Calcul du taux de croissance de la valeur ajoutée pour chaque entreprise du groupe

```python
for i in range(0,7):
    valeurAjout1 = "valeur_ajout_"+str(groupes[i])
    valeurAjout2 = "valeur_ajout_"+str(groupes[i+1])
    valeurAjout3 = "valeur_ajout_"+str(groupes[i+2]) 
    (df[i])["TCAM"] = ((((df[i])[valeurAjout3])/((df[i])[valeurAjout1]))**(1/3) - 1)*100
    (df[i])["Groupe"] = "groupe "+str(i+1)
    nombreEmployes1 = "eff_moyen_personnel_"+str(groupes[i])
    nombreEmployes2 = "eff_moyen_personnel_"+str(groupes[i+1])
    nombreEmployes3 = "eff_moyen_personnel_"+str(groupes[i+2])
    
    
    conditions  = [(df[i])["TCAM"] < 20 ,\
                   ((df[i])["TCAM"]>=20) & ((df[i])[nombreEmployes1] >= 10) & ((df[i])[nombreEmployes2] >= 10) & ((df[i])[nombreEmployes3] >= 10),\
                    ((df[i])["TCAM"]>=20) & (((df[i])[nombreEmployes1] < 10) | ((df[i])[nombreEmployes2] < 10) | ((df[i])[nombreEmployes3] < 10))]
    choices     = [ "Normal","EFC","Quasi EFC"]
    
    (df[i])["classe"] = np.select(conditions, choices, default=np.nan)
    
    
   
```

## Concaténation de tous les groupes en un seul dataframe

```python

df_all = df[0]
for i in range(1,7):
    df_all = df_all.append(df[i],ignore_index=True)
print(df_all.shape)
df_all.head()
```
